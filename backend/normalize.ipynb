```python
import pandas as pd
import re
from rapidfuzz import fuzz, process

# Load dataset (Update path if necessary)
df = pd.read_csv('backend/cleaned.csv')

# Normalize text function (remove special characters & convert to lowercase)
def normalize_text(text):
    if pd.isna(text):  # Handle NaN values
        return ""
    text = str(text).lower().strip()
    text = re.sub(r'[^a-z\s,]', '', text)  # Keep only letters and commas
    return text

# Define standard known ingredients (expand as needed)
known_ingredients = [
    "milk", "almond milk", "soy milk", "sugar", "butter", "flour", "eggs", "salt",
    "pepper", "cheese", "tomato", "onion", "garlic", "chicken", "beef", "pork", 
    "olive oil", "honey", "vanilla", "cinnamon", "mustard", "mayonnaise", "bread", 
    "yogurt", "cream", "parsley", "cilantro", "thyme", "oregano", "bay leaves", "carrot",
    "mushrooms", "coconut milk", "spinach", "cumin", "vinegar", "eggplant", "chili",
    "lettuce", "radish", "turmeric", "ginger", "sesame", "soy sauce", "lemon", "lime"
]

# Function to extract only relevant ingredients
def extract_clean_ingredients(ingredient_list):
    if pd.isna(ingredient_list) or not isinstance(ingredient_list, str):
        return "No Match Found"  # Handle empty values

    normalized_ingredients = normalize_text(ingredient_list)  # Normalize text
    ingredients = normalized_ingredients.split(',')  # Split ingredients list
    
    clean_ingredients = set()  # Use a set to avoid duplicates
    for ingredient in ingredients:
        ingredient = ingredient.strip()
        match, score, _ = process.extractOne(ingredient, known_ingredients, scorer=fuzz.partial_ratio)
        if score > 75:  # Lowered threshold for better recognition
            clean_ingredients.add(match)  # Add to set to prevent duplicates
            
    return ', '.join(clean_ingredients) if clean_ingredients else "No Match Found"

# Apply the function to extract cleaned ingredients
df['Cleaned_Ingredients'] = df['Ingredients'].apply(extract_clean_ingredients)

# Save the updated dataset with the new column
cleaned_file_path = "backend/cleaned_with_ingredients.csv"
df.to_csv(cleaned_file_path, index=False)

# Provide the file path for download
cleaned_file_path
```
